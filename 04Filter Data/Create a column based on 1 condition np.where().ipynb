```python
import pandas as pd
```

```python
df_laptop = pd.read_csv('laptop_price.csv',encoding='unicode_escape')
```

1. create a column based on 1 condition:
np.where()

```python
import numpy as np
```

```python
# where > condition : yes : no
np.where( df_laptop['Price_euros'] > 2000 , 'Expensive', 'Cheap')
```

```python
df_laptop['Price-tier'] = np.where( df_laptop['Price_euros'] > 2000 , 'Expensive', 'Cheap')
```

```python
df_laptop.head()
```

```python
#count value in price-tier column
df_laptop.value_counts('Price-tier')
```
