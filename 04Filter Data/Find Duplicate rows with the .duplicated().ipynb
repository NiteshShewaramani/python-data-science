Example 2
find the cheapest and most expensive laptop of each company using sort_values() and duplicated() method
(keep "first","last" and False)

```python
#sort dataframe by company and price
df_laptops = df_laptops.sort_values(by=['Company','Price_euros'])
#override the dataframe
```

```python
#check all values categories in Company column
df_laptops.value_counts('Company')
```

```python
#duplicated values in company column  (default -> keep ='first')
duplicated_first = df_laptops.duplicated('Company',keep='first') 
```

```python
df_laptops[duplicated_first]
```

```python
df_laptops[~duplicated_first] #not duplicate value
```

```python
df_laptops[~duplicated_first][['Company','Price_euros']] #not duplicate value also showing cheapest laptop in each company
```

```python
#duplicated values in company column  (default -> keep ='first')
duplicated_last = df_laptops.duplicated('Company',keep='last') #using last now 
```

```python
#keep='last' most expensive laptop per company
df_laptops[~duplicated_last] #not duplicate value also showing expensive laptop in each company
```

```python
df_laptops[~duplicated_last][['Company','Price_euros']] #not duplicate value also showing expensive laptop in each company
```

```python
#duplicated values in company column  (default -> keep ='first')
duplicated_false = df_laptops.duplicated('Company',keep=False) #using False now
```

```python
#keep=False 
df_laptops[~duplicated_false] # we are not keeping first or last duplicate values
```
