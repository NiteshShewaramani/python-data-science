```python
import pandas as pd

df_laptops = pd.read_csv('laptop_price.csv',encoding= 'unicode_escape')
```

1 drop_duplicates()

```python
#drop duplicates in 2 or more columns
df_laptops.drop_duplicates(['Company'])
```

```python
df_laptops.drop_duplicates(['Company'])[['Company']]
```

```python
df_laptops = df_laptops.sort_values(by=['Company','Price_euros'])
```

```python
#cheapest keep='first' default
df_laptops.drop_duplicates(['Company'],keep='first')
```

```python
df_laptops.drop_duplicates(['Company'],keep='first')[['Company','Price_euros']]
```

```python
#show most expensive laptop of each company
df_laptops.drop_duplicates(['Company'],keep='last')[['Company','Price_euros']]
```

```python
#ignore the index 
df_laptops.drop_duplicates(['Company'],keep='last',ignore_index=True)
```
