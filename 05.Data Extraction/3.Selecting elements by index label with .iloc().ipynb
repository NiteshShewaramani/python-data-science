```python
import pandas as pd

#read csv file
df = pd.read_csv('players_20.csv' , encoding = 'unicode_escape')

#set index
#df = df.set_index('short_name') #creates a copy so we need to overwrite  
df.set_index('short_name',inplace=True)

#select columns
df = df[['long_name',	'age',	'dob',	'height_cm',	'weight_kg',	'nationality',	'club']]

#show dataframe
df
```

1. selecting with a single value
iloc[row_position,column_position]


```python
# get the height of messi

df.iloc[0,3]
```

```python
#get weight of ronaldo
df.iloc[1,4]
```

```python
#get all rows inside the height_cm column 

df.iloc[:,3] # : means get all
```

```python
#get all column values for messi

df.iloc[0,:]# : means get all
```

2 Selecting with a list of values

```python
#get all data for messi and ronaldo
#here last value is exclusive
df.iloc[[0,2] ,:]
```

```python
#get the height of messi and ronaldo

df.iloc[[0,2] ,3]
```

```python
#get height and weight of messi and ronaldo
#use index no , index start from 0
df.iloc[[0,2] , [3,4] ]
```

3. selecring a range of data with a slice

start:stop (note start is included but stop is exclusive)

```python
#slice columns labels age/club
#player = messi and ronaldo

df.iloc[[0,1] , 1:6+1 ]
```

```python
#slice index labels (top 1 to top10 player)
#columns - age to weight

df.iloc[0:11,1:5]
```

4. selecting with conditions

iloc cannot accept boolean series , but only a boolean list , we have to use list() function to convert series to boolean list


```python
#one condition : select player with height above 180cm
#columns age to weight

# df['height_cm']>180 condition

df.iloc[list(df['height_cm']>180) , 1:5]
```

```python
#multiple condition select player  with height above 180cm and player are from argentina

# df['height_cm']>180 condition
# df['nationality'] == "Argentina" condition2

conditions =  list((df['height_cm']>180)  & (df['nationality'] == "Argentina")) 

df.iloc[conditions , :]

```
