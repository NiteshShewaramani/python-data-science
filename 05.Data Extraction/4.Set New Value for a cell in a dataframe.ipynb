```python
import pandas as pd

#read csv file
df = pd.read_csv('players_20.csv' , encoding = 'unicode_escape')

#set index
#df = df.set_index('short_name') #creates a copy so we need to overwrite  
df.set_index('short_name',inplace=True)

#select columns
df = df[['long_name',	'age',	'dob',	'height_cm',	'weight_kg',	'nationality',	'club']]

#show dataframe
df
```

1 Set value to one cell

```python
#update messi height

df.loc['L. Messi','height_cm'] = 175
```

```python
# show df
df
```

2 set value to entire column

```python
#set height of all player to 190
# get all rows df.loc[:,'height_cm'] 
df.loc[:,'height_cm'] = 190
#show df
df
```

3 set value to an entire row 

```python
#get all columns that correspond to player ranked last in fifa
#-1 for last value
df.iloc[-1,:]
```

```python
# set null values to all cols that belong to last player
import numpy as np

np.nan #null value

df.iloc[-1,:] = np.nan
```

```python
#show df
df
```

4. set value to multiple cells
5. 

```python
#set value for all items matching the list of labels

players = ['L. Messi','Cristiano Ronaldo']

df.loc[players , 'height_cm'] = 175
```

```python
df
```

5. set value for rows matching a condition

```python
#set value for rows matching a condition
col = ['age',	'dob',	'height_cm',	'weight_kg']

df['height_cm']>180

df.loc[ df['height_cm']>180 , col] = 0
```
