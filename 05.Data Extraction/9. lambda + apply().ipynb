```python
import pandas as pd

#read csv file
df = pd.read_csv('players_20.csv' , encoding = 'unicode_escape')

#set index
#df = df.set_index('short_name') #creates a copy so we need to overwrite  
df.set_index('short_name',inplace=True)

#select columns
df = df[['long_name',	'age',	'dob',	'height_cm',	'weight_kg',	'nationality',	'club']]

#show dataframe
df
```

1. Lambda function

```python
#basic function
def sum_values(a,b):
    x=a+b
    return x
sum_values(2,3)
```

```python
sum_value_lambda = lambda a,b : a+b

sum_value_lambda(2,5)
```

2 Apply + Lambda function

```python
#use lambda func to convert height to m

df['height_cm'].apply(lambda x :  x/100)
```

```python
#alternative
df['height_cm']/100
```

```python
#use lambda to convert long_name to upper case

df['long_name'].apply(lambda x : x.upper())
```

```python
# use lambda to get the year or dob series
df['dob'] = df['dob'].astype('datetime64[ns]')

df['dob'].apply(lambda x:x.year)
#alternative
df['dob'].dt.year
```

```python
#use lambda to get bmi of the dataframe
df.apply(lambda row : row['weight_kg']/ (row['height_cm']/100)**2 , axis=1) #axis 1 as we get multiple cols
```
