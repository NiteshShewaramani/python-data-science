```python
import pandas as pd


df = pd.read_csv('Car_sales.csv')

df
```

```python
#selecting columns
df_filter = df[['Manufacturer','Sales_in_thousands','Vehicle_type','Price_in_thousands','Engine_size','Horsepower','Fuel_capacity']]
df_filter
```

1 agg()

```python
#calculate the sum of cols 
df_filter.agg('sum')
```

```python
#calculate the mean of numeric values
df_filter[['Sales_in_thousands','Price_in_thousands','Engine_size']].agg('mean')
```

```python
#cal mean and count values
df_filter[['Sales_in_thousands','Price_in_thousands','Engine_size','Fuel_capacity']].apply(['mean','count'])
```

```python
#different aggregations per col : cal sum and mean of 'Sales_in_thousand' and sum and max for Price_in_thousands


df_filter.agg({'Sales_in_thousands':['sum','mean'],
               'Price_in_thousands':['sum','max'] })
```

```python
#aggregate over cols
df_filter[['Sales_in_thousands','Price_in_thousands']].agg('sum',axis=1) #takes null as 0
```

```python
#equivalent (NaN behaves differently)

df_filter['Sales_in_thousands'] + df_filter['Price_in_thousands']
```

```python
#aggregate diff func and rename the index of the resulting dataframe

df_filter.agg( x  = ('Sales_in_thousands','sum'),
               y  = ('Price_in_thousands','sum')  )
```

2 the split-apply combine strategy 

```python
#split data into seperate groups

passenger_filter  = df_filter['Vehicle_type']=="Passenger"
car_filter  = df_filter['Vehicle_type']=="Car"

```

```python
#apply an operation

passsenger_average  = df[passenger_filter]['Sales_in_thousands'].mean()

car_average  = df[car_filter]['Sales_in_thousands'].mean()
```

```python
#combine results

pd.DataFrame({'Vehicle_type':['Car','Passenger'],
              'Sales_in_thousands' :[car_average , passsenger_average] } )
```

4 Groupby() and agg()

```python
#find the min and max val on each col

df_filter.groupby('Vehicle_type').min()

df_filter.groupby('Vehicle_type').max()

df_filter.groupby('Vehicle_type').agg(['min','max'])
```

```python
#set a name for aggregated col
#find the min Engine_size and max horsepower

df_filter.groupby('Vehicle_type').agg(  min_engine_size = ('Engine_size','min') , 
                                        max_horsepower  = ('Horsepower','max') )
```

```python
#cal the sum of Sales_in_thousands and mean of Price_in_thousands group by Manufacturer
df_filter.groupby('Manufacturer').agg(  sum_sales = ('Sales_in_thousands','sum') , 
                                        mean_price  = ('Price_in_thousands','mean') )
```

```python
df_filter.groupby('Manufacturer').sum()['Sales_in_thousands']
```

5 Lambda

```python
df_filter.groupby('Manufacturer').sum().head() 
```

```python
#get actual sales and price values
#lambda x:x*1000 -> returns a number multiplied by 1000
df_filter.groupby('Manufacturer').sum().apply(lambda x:x*1000)[['Sales_in_thousands','Price_in_thousands']]
```

```python
#subtract values from the mean of each group 
#x is value in dataframe x.mean() = average values of each group like ('Bwm','audi' , group by is manufacturer)
df_filter[['Sales_in_thousands','Price_in_thousands','Manufacturer']].groupby('Manufacturer').apply(lambda x : x - x.mean())
```

6 The filter() method 

```python
def filter_func(x):
    #x represent a dataframe
    #52 is mean of Sales_in_thousands column
    return x['Sales_in_thousands'].sum() > 52
#we are filtering values where sum of group(ex - bmw , audi) is less than mean value of whole dataset
df_filter.groupby('Manufacturer').filter(filter_func) #porche bmw  jaguar saab audi infiniti got filtered out
```
