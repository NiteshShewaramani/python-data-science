```python
import pandas as pd

df_movies_org = pd.read_csv('IMDb movies.csv', low_memory=False)
df_ratings_org = pd.read_csv('IMDb ratings.csv', low_memory=False)
```

```python
df_movies = df_movies_org[['imdb_title_id','title','year','genre','country']]

df_ratings = df_ratings_org[['imdb_title_id','total_votes','mean_vote']]
```

1 concat()
1.1 concatenate vertically

```python


df1 = pd.DataFrame({'id':['A','B','C','D'] ,
                     'age':[30,23,25,22] })
df2 = pd.DataFrame({'id':['E','F','G','H'] ,
                     'age':[40,21,19,24] })
```

```python
pd.concat([df1 , df2], axis = 0, ignore_index=True) #we want o concatenate vertically
```

Joins 
1 Merge()

```python
df1 = pd.DataFrame({'id':['A','B','C','D'] ,
                     'age':[30,23,25,22] })
df2 = pd.DataFrame({'id':['C','D','E','F'] , 
                    'job':['doctor','statistician','accountant','developer'] })
```

1.1 Inner join

```python
df_inner_join = df1.merge(df2, on='id' , how='inner')
df_inner_join
```

```python
print(df1.shape)
print(df2.shape)
print(df_inner_join.shape)
```

1.1.1 Exercise

```python
#merge df_movies and df_raatings 

df_movies.merge(df_ratings , on='imdb_title_id' , how='inner')
```

1.2 Outer join (Full join)

```python
df1.merge(df2 , on='id' , how='outer')
```

1.2.1 Exercise
#merge movies and ratings df outer join


```python
df_movies.merge(df_ratings , on='imdb_title_id' , how='outer')
```

Joins (Exclusive full join) - does not include common values between 2 df

```python
df1.merge(df2 , on='id' , how='outer', indicator=True)
```

```python
df1.merge(df2 , on='id' , how='outer'
          , indicator=True).query('_merge=="left_only" or _merge=="right_only"')
```

1.3.1 Exercise

```python
df_movies.merge(df_ratings , on='imdb_title_id' , how='outer' , indicator=True)
```

```python
df_movies.merge(df_ratings , on='imdb_title_id' , how='outer' , indicator=True).query('_merge=="left_only" or _merge=="right_only"')
```

1.4 Joins Left join - only gets data from left table 

```python
df1.merge(df2 , on='id' , how='left')
```

1.5 Right join - only gets data from right table 

```python
df1.merge(df2 , on='id' , how='right')
```

1.6 Exclusive left join only - data is only picked from the first table and also which is not common with second table

```python
df1.merge(df2 , on='id' , how='outer'
          , indicator=True).query('_merge=="left_only"')
```

1.7 Exclusive right join only - data is only picked from the second table and also which is not common with first table

```python
df1.merge(df2 , on='id' , how='outer'
          , indicator=True).query('_merge=="right_only"')
```
