1 The re module 

```python
import re
```

1.1 re.search()

```python
text = """Hello World.
I love Python!!!!
My first lucky number 987-654-321
My second lucky number 876-543-210
My first lucky number 765-432-100
My first lucky number 123-456-123-123-456
"""
```

```python
re.search(r'\d',text)
```

```python
# flags

re.search(r'^hello', text, flags=re.I) #I means case insencetive or ignore case
```

2 re.findall()

```python
re.findall(r'\d', text) #find all digits inside text
```

```python
#find all puctuations !
re.findall(r'[^\w\s]', text) #find all digits inside text
```

```python
#match right date format (dd-mm-yyyy)
text= '''
13-04-2021
2021-13-04
2021-04-12
'''
re.findall(r'\d{2}-\d{2}-\d{4}', text)
```

```python
#match right username format (4-14 charc and only letters or digit)
text='''
username10
abc
10
'''
re.findall(r'[a-zA-Z0-9]{4,14}', text)
```
