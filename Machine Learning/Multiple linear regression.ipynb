```python
import pandas as pd

df_boston = pd.read_csv('Boston House Prices.csv')
```

```python
import statsmodels.api as sm
```

```python
#dependent variable
y = df_boston['Value']

#independent variable 

X = df_boston[['Rooms','Distance']]
```

```python
#adding a constant
X = sm.add_constant(X) #adding a constant
#fitting the model
lm = sm.OLS(y, X).fit() #(dependent variable , independent variable)
```

```python
lm.summary()
```

```python
from sklearn import linear_model
```

```python
#dependent variable
y = df_boston['Value']

#independent variable 

X = df_boston[['Rooms','Distance']]
```

```python
#Fitting the model
lm = linear_model.LinearRegression()
lm.fit(X , y ) #order has changes (independent , dependent)
```

```python
lm.predict(X)
```

```python
lm.predict(X)[0:5]
```

Regression Table 

```python
#r squared score
lm.score(X , y)
```

```python
# coefficient

lm.coef_
```

```python
#intercept
lm.intercept_
```
