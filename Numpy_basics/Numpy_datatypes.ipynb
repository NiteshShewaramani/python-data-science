```python
import numpy as np

print(np.__version__)
```

```python
#create array

np_array = np.array([1,2,3,4])
print(np_array)
mathsOperation = np_array * 2 #vectorized operation

print(mathsOperation)

print(type(mathsOperation))
```

```python
#dtype = tells numpy what kind of values are stored in an array otherwise numpy guesses 
#integer (int8,int16,int32,int64)

np_array = np.array([1,2,3,4,5])
print(np_array)
print(np_array.dtype)
print(f"{np_array.nbytes} bytes")
```

```python
np_array = np.array([1,2,3,4,5], dtype = np.int32)
print(np_array)
print(np_array.dtype)
print(f"{np_array.nbytes} bytes")
```

```python
np_array = np.array([1,2,3,4,5], dtype = np.int16)
print(np_array)
print(np_array.dtype)
print(f"{np_array.nbytes} bytes")
```

```python
np_array = np.array([1,2,3,4,5], dtype = np.float16)
print(np_array)
print(np_array.dtype)
print(f"{np_array.nbytes} bytes")
```

```python
np_array = np.array([1,2,3,4,5,0], dtype = np.bool_) #bool_ is numpy boolean 
#any non zero is true
print(np_array)
print(np_array.dtype)
print(f"{np_array.nbytes} bytes")
```

```python
np_array = np.array([1,2,3,4,5], dtype = np.str_) #str_ is numpy string 
#any non zero is true
print(np_array)
print(np_array.dtype)
print(f"{np_array.nbytes} bytes")
```

```python
np_array = np.array(["apple","orange","banana"], dtype = np.str_) #str_ is numpy string 
#any non zero is true
print(np_array)
print(np_array.dtype)
print(f"{np_array.nbytes} bytes")
```

```python
np_array = np.array(["apple","orange","banana"], dtype = "<U4") #str_ is numpy string 
#any non zero is true
print(np_array)
print(np_array.dtype)
print(f"{np_array.nbytes} bytes")
```

```python
#objects

np_array = np.array(["apple",1.2,"banana"], dtype = np.object_) #object_ is python object so speed is slow
#any non zero is true
print(np_array)
print(np_array.dtype)
print(f"{np_array.nbytes} bytes")
```

```python
np_array = np.array([1,2,3,4,5]) 
#any non zero is true
np_array = np_array.astype(np.float16)
print(np_array)
print(np_array.dtype)
print(f"{np_array.nbytes} bytes")
```
