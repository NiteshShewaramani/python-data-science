```python
import numpy as np # linear algebra


rng = np.random.default_rng()

print(rng.integers(low=1, high=7)) # random no btw 1 and 6
```

```python
print(rng.integers(low=1, high=100 , size = 3)) #3 random no btw 1 and 100
```

```python
print(rng.integers(low=1, high=100 , size = (3,2))) #3,2 random no btw 1 and 100
```

```python
print(np.random.uniform()) # random floating point between 0 and 1
```

```python
array = np.array([1,2,3,4,5])

print(array)

rng.shuffle(array)
print(array)
```

```python
fruits = np.array(['apple','orange','banana','coconut','pineapple','kiwi'])

rng.choice(fruits)
```

```python
fruits = np.array(['apple','orange','banana','coconut','pineapple','kiwi'])

rng.choice(fruits , size=(2,3))
```
