```python
import numpy as np # linear algebra

#scaler arithmetic 
array = np.array([1,2,3])

print(array+1)
print(array-2)

print(array*3)
print(array / 4)
print(array**5)
```

```python
#vectorized math function 
array = np.array([1,2,3])

print(np.sqrt(array))
```

```python
array = np.array([1.2,2.5,3.7])
print(np.floor(array))
print(np.ceil(array))
print(np.round(array))
```

```python
radius = np.array([1,2,3])

print(np.pi * radius**2) 
```

```python
array1 = np.array([1,2,3])
array2 = np.array([4,5,6])

print(array1+array2)

print(array1-array2)

print(array1*array2)

print(array1/array2)

print(array1**array2)
```

```python
#comparision operator 
scores = np.array([91,55,100,73,82,64])

print(scores >= 60)
```

```python
scores = np.array([91,55,100,73,82,64])

print(scores < 60)
```

```python
#scores < 60 filter condition
scores[scores < 60] = 0
print(scores)
```
