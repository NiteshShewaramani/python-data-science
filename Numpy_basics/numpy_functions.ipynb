```python
import numpy as np # linear algebra

array = np.zeros(10)
print(array)
```

```python
array = np.zeros((2,10))
print(array)
```

```python
array = np.zeros((2,3,10))
print(array)
```

```python
array = np.ones((2,3,10))
print(array)
```

```python
array = np.full((2,3,10) , 9) #shape in tuple , then value as another parameter
print(array)
```

```python
array = np.eye(2) #identity matrix
print(array)
```

```python
array = np.eye(4) #identity matrix
print(array)
```

```python
array = np.empty((2,3)) # garbage values or random values
print(array)
```

```python
array = np.arange(0,100) # start , stop , step
print(array)
```

```python
array = np.arange(0,100,2) # start , stop , step
print(array)
```

```python
array = np.arange(0,50,0.1) # start , stop , step
print(array)
```

```python
array = np.linspace(0,10,3) # start , stop , num
print(array)
```

```python
array = np.linspace(0,10,4) # start , stop , num
print(array)
```

```python
array = np.linspace(0,10,5) # start , stop , num
print(array)
```
