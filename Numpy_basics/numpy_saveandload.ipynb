```python
import numpy as np # linear algebra

array = np.array([[1,2,3],[4,5,6]])

np.save("data_numpy_array",array)

print("numpy array was saved")
```

```python
array1 = np.array([[1,2,3],[4,5,6]])
array2 = np.array([1,2,3,7,8,9])

np.savez("save_multiple_array", array1,array2)#save multiple array

print("numpy array was saved")
```

```python
array1 = np.array([[1,2,3],[4,5,6]])
array2 = np.array([1,2,3,7,8,9])

np.savez_compressed("save_multiple_array_compressed", array1,array2)#save multiple array

print("numpy array was saved")
```
