Adding a new column

assign()
add muliple columns in a single line of code
when you need to overwrite values of existing columns (best practice)
it returns a new object(a copy) with all the original columns in addition to new ones


```python
score1 = np.random.randint(1,100,size=1000)
score2 = np.random.randint(1,100,size=1000)

#create series now using the above random numbers
series1 = pd.Series(score1, index=np.arange(0,1000))
series2 = pd.Series(score2, index=np.arange(0,1000))
#using assign to add multiple columns
df_exams.assign(newcol1=series1,newcol2=series2) #copy only not the original 

#df_exams = df_exams.assign(newcol1=series1,newcol2=series2) #updates the original
```

insert()

insert a new column at a specific position or index
does not create a copy , updates the original

```python
#using index to add a column
df_exams.insert(1,'test',series1)
```

Sort a dataframe

```python
#sort by 1 column
#creates a copy of dataframe does not update the original
df_exams.sort_values(by='math score')
```

```python
#sort by 1 column descending

df_exams.sort_values(by='math score',ascending = False)
```

```python
#sort by multiple columns descending

df_exams.sort_values(by=['math score','reading score'],ascending = False)
```

```python
#inplace = True means update the original data frame not the copy
df_exams.sort_values(by=['math score','reading score'],ascending = False,inplace=True)
```

```python
df_exams
```

```python
#sort descending with a key function
df_exams.sort_values(by=['race/ethnicity'],ascending = True, key = lambda col : col.str.lower() ) #get a string value and get lower case
```
