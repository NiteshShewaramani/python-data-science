2 Read a .csv file from Multiple URLs with pandas

LINK : root + season + league + .csv

```python
# https://www.football-data.co.uk/mmz4281/2526/E0.csv
# https://www.football-data.co.uk/mmz4281/2526/E1.csv
# https://www.football-data.co.uk/mmz4281/2526/E2.csv
# https://www.football-data.co.uk/mmz4281/2526/E3.csv
# https://www.football-data.co.uk/mmz4281/2526/EC.csv

#create a root variable 
root = 'https://www.football-data.co.uk/mmz4281/'
season = '2526/'

#creating a list of leagues
leagues = ['E0','E1','E2','E3','EC']
frames  = [] #list of dataframes
for league in leagues:
    df = pd.read_csv(root + season + league + '.csv' , encoding='unicode_escape')
    frames.append(df)
    
```

```python
# https://www.football-data.co.uk/mmz4281/2526/E0.csv
# https://www.football-data.co.uk/mmz4281/2526/E1.csv
# https://www.football-data.co.uk/mmz4281/2526/E2.csv
# https://www.football-data.co.uk/mmz4281/2526/E3.csv
# https://www.football-data.co.uk/mmz4281/2526/EC.csv

#create a root variable 
root = 'https://www.football-data.co.uk/mmz4281/'

#creating a list of leagues
leagues = ['E0','E1','E2']
frames  = [] #list of dataframes
for league in leagues:
    for season in range(20,26):
        
        df = pd.read_csv(root + str(season) + str(season+1)  + '/' + league + '.csv' , encoding='unicode_escape')
        #add a new column indicating which season is in dataframe
        df.insert(1,'season', str(season))
        frames.append(df)

```

Organize all the data in a dictionary

```python
#create a dictionary with original name of league as key
football_country_codes = {
    'Spanish La Liga':'SP1',
    'German Bundesliga':'D1',
    'English Premier League':'E0',
    'English League 1':'E2'
}


#create a root variable 
root = 'https://www.football-data.co.uk/mmz4281/'
dict_historic_data = {}


for league in football_country_codes:
    frames  = [] #list of dataframes
    for season in range(24,26):
        
        df = pd.read_csv(root + str(season) + str(season+1)  + '/' + football_country_codes[league] + '.csv' , encoding='unicode_escape')
        #add a new column indicating which season is in dataframe
        df.insert(1,'season', str(season))
        frames.append(df)
    #concatenate all seasons in 1 dataframe
    df_concat =  pd.concat(frames)
    #combine league name(key) with dataframe of all seasons(24-25) 
    dict_historic_data[league] = df_concat
```
