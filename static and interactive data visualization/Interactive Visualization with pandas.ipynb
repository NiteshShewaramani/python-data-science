```python
import pandas as pd
import cufflinks as cf
from IPython.display import display, HTML
import plotly.express as px
cf.set_config_file(sharing='public', theme='ggplot', offline=True)

cf.getThemes()
```

```python
df = pd.read_csv('population_total.csv')
#read dataframe
df.dropna(inplace=True)

df_pivot = df.pivot(index='year', columns= 'country' ,values='population')
df_pivot_filter = df_pivot[['United States','India','China','Brazil','Indonesia']]

```

1 Line Plot

```python
fig = px.line(df_pivot_filter,labels={
        "index": "Year",
        "value": "Population",
        "variable": "Country"
    })
fig.show()
```

2 Barplot

```python
#selecting only one year (2020)
df_pivot_2020 = df_pivot_filter [df_pivot_filter.index.isin([2020.0])]

# transpose data from rows to col and viceversa
df_pivot_2020T = df_pivot_2020.T

#show dataframe
df_pivot_2020T
```

```python
fig = px.bar(
    df_pivot_2020T,
    title="Population 2020",
    labels={
        "index": "Year",
        "value": "Population",
        "variable": "Country"
    }
)

fig.show()
```

```python
#selecting a few years 

df_pivot_sample = df_pivot_filter [df_pivot_filter.index.isin([1980 ,1990 , 2000,2010, 2020])]
df_pivot_sample

fig = px.bar(
    df_pivot_sample,
    title="Population 1980-2020",
    labels={
        "index": "Year",
        "value": "Population",
        "variable": "Country"
    }
)

fig.show()
```

3 Piechart

```python
df_pivot_2020T = df_pivot_2020.T
df_pivot_2020T
```

```python
fig = px.pie(
    df_pivot_2020T,
    names=df_pivot_2020T.index,
    values=2020.0,
    title="Population Distribution - 2020"
)

fig.show()
```
